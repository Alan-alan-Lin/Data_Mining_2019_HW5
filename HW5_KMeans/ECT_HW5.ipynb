```python
#讀csv檔
import pandas as pd
data = pd.read_csv("BreastCancer.csv")
data
```

```python
#切分input和output
#x作為input
#y作為output
#特徵欄位
feature = data.loc[:,['radius_mean','area_mean']]
#目標欄位
target = data.loc[:,['diagnosis']]
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=0).fit(feature)
```

```python
kmeans_predict = kmeans.fit_predict(feature)
```

```python
#繪圖
import matplotlib.pyplot as plt
x = data.loc[:,'radius_mean']
y = data.loc[:,'area_mean']
plt.scatter(x, y, c = kmeans_predict)
```
